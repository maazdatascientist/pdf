#PRE-PROCESIING THE TEXT OF THE PDF

```python
import re
import nltk
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')


# Download the stopwords if needed
nltk.download('stopwords')

# Define the set of stopwords to use
stop_words = set(stopwords.words('english'))

def preprocess_text(text1):
    # Remove non-alphanumeric characters and extra whitespaces
    text1 = re.sub(r'[^\w\s]', ' ', text1)
    print('After regex substitution:', text1)
    text1 = re.sub(r'\s+', ' ', text1)

    # Lowercase the text
    text1 = text1.lower()
    print('After lowercase:', text1)

    # Tokenize the text
    tokens = nltk.word_tokenize(text1)
    print('Tokenized text:', tokens)

    # Remove stopwords
    tokens = [token for token in tokens if token not in stop_words]
    print('After stopword removal:', tokens)

    # Join the tokens back into text1
    text1 = ' '.join(tokens)
    print('Final text:', text1)

    return text1.strip()
```

# MODEL

```python
!pip install numpy datasets transformers
!pip install datasets
import numpy as np
from datasets import load_dataset, Dataset
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, TrainingArguments, Trainer, default_data_collator

```

```python
# Load SQuAD dataset
squad = load_dataset("squad")
squad_train = squad["train"].shuffle(seed=42).select(range(5000))

# Split train dataset into train and validation datasets
train_size = int(0.8 * len(squad_train))
train_dataset = squad_train[:train_size]
val_dataset = squad_train[train_size:]

# Convert train and validation datasets to datasets.Dataset objects
train_dataset = Dataset.from_dict(train_dataset)
val_dataset = Dataset.from_dict(val_dataset)
```

```python
# Apply preprocessing to train and validation datasets
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-distilled-squad")
```

```python
# Define the preprocessing function
def preprocess_function(examples):
    questions = [q.strip() for q in examples["question"]]
    inputs = tokenizer(
        questions,
        examples["context"],
        max_length=384,
        truncation="only_second",
        return_offsets_mapping=True,
        padding="max_length",
    )

    offset_mapping = inputs.pop("offset_mapping")
    answers = examples["answers"]
    start_positions = []
    end_positions = []

    for i, offset in enumerate(offset_mapping):
        answer = answers[i]
        start_char = answer["answer_start"][0]
        end_char = answer["answer_start"][0] + len(answer["text"][0])
        sequence_ids = inputs.sequence_ids(i)

        # Find the start and end of the context
        idx = 0
        while sequence_ids[idx] != 1:
            idx += 1
        context_start = idx
        while sequence_ids[idx] == 1:
            idx += 1
        context_end = idx - 1

        # If the answer is not fully inside the context, label it (0, 0)
        if offset[context_start][0] > end_char or offset[context_end][1] < start_char:
            start_positions.append(0)
            end_positions.append(0)
        else:
            # Otherwise it's the start and end token positions
            idx = context_start
            while idx <= context_end and offset[idx][0] <= start_char:
                idx += 1
            start_positions.append(idx - 1)

            idx = context_end
            while idx >= context_start and offset[idx][1] >= end_char:
                idx -= 1
            end_positions.append(idx + 1)

    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions
    return inputs
```

```python
# Apply preprocessing to train and validation datasets
tokenized_train_dataset = train_dataset.map(preprocess_function, batched=True)
tokenized_val_dataset = val_dataset.map(preprocess_function, batched=True)
```

```python
# Freeze the DistilBERT base model
model = AutoModelForQuestionAnswering.from_pretrained("distilbert-base-uncased-distilled-squad", return_dict=True)
for param in model.distilbert.parameters():
    param.requires_grad = False
```

```python
# Define the training arguments
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="steps",
    eval_steps=500,
    save_total_limit=2,
    learning_rate=5e-5,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    num_train_epochs=15,
    weight_decay=0.01,
    push_to_hub=False,
    logging_dir='./logs',
    logging_steps=500,
    fp16=True,  # Use mixed precision training
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
)
```

```python
# Define a function to compute the accuracy and loss during training
import numpy as np

def compute_metrics(pred):
    labels = pred.label_ids
    preds = np.array(pred.predictions).argmax(-1)
    accuracy = (preds == labels).astype(np.float32).mean().item()
    return {"accuracy": accuracy}
```

```python
# Create the Trainer object and train the model
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,
    eval_dataset=tokenized_val_dataset,
    data_collator=default_data_collator,
    compute_metrics=compute_metrics,)
```

```python

# Train the model
trainer.train()
```

```python
from transformers import pipeline

# Define the question answering pipeline
question_answering = pipeline(
    "question-answering",
    model=model,
    tokenizer=tokenizer,
)
import torch
from transformers import AutoTokenizer, AutoModelForQuestionAnswering



# move model to GPU
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model.to(device)

# set context and question
context = "The term 'machine learning' was coined by Arthur Samuel in 1959. Machine learning is a subset of artificial intelligence. It is the study of computer algorithms that improve automatically through experience."
question = "who coined the term 'machine learning'?"

# tokenize inputs and move to GPU
inputs = tokenizer.encode_plus(question, context, return_tensors="pt")
for key in inputs:
    inputs[key] = inputs[key].to(device)

# get answer
with torch.no_grad():
    outputs = model(**inputs)
answer_start_scores = outputs.start_logits
answer_end_scores = outputs.end_logits
start_idx = torch.argmax(answer_start_scores)
end_idx = torch.argmax(answer_end_scores) + 1
answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][start_idx:end_idx]))
print(answer)


#handling Expected all tensors to be on the same device, but found at least two devices, cuda:0 and cpu! (when checking argument for argument index in method wrapper__index_select)
```
